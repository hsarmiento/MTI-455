# Extraer y mostrar el contenido de las noticias más vistas de emol

```python
context = ssl._create_unverified_context()

# Aquí ingresamos a la página principal de EMOL
page = urlopen('http://www.emol.com', context=context).read()
soup = BeautifulSoup(page)

# Ubicamos el div que contiene las noticias más vistas
mostViewed = soup.findAll("div",
                         {"class": "caja_contenedor_masvistos_modulo_texto"})

# Obtenemos las URLs de aquellas noticias
print("###")
urlViewed = []


for div in mostViewed:
    if div.a:
        urlViewed.append(div.a["href"])


# Para cada una de esas noticias ingresamos a su página
# y obtenemos su título y su contenido.
count = 1
for url in urlViewed:
    print("### Noticia {}: ".format(count))
    try:
        page = urlopen(
            "http://www.emol.com{}".format(url), context=context
        ).read()
        soup = BeautifulSoup(page)
    except:
        print("Error al intentar recuperar noticia.")


    print("Título: ")
    try:
        title = soup.find("h1",
                         {"id": "cuDetalle_cuTitular_tituloNoticia"})
        print(title.getText())
    except:
        print("Error al intentar recuperar título de noticia.")


    print("Contenido: ")
    try:
        content = soup.find("div",
                         {"id": "cuDetalle_cuTexto_textoNoticia"})
        print(content.getText())
    except:
        print("Error al intentar recuperar contenido de noticia.")
    print("")
    count += 1
```

## Lo mismo que lo anterior, pero guardando el contenido en un archivo tabulado

```python
context = ssl._create_unverified_context()

# Aquí ingresamos a la página principal de EMOL
page = urlopen('http://www.emol.com', context=context).read()
soup = BeautifulSoup(page)

# Ubicamos el div que contiene las noticias más vistas
mostViewed = soup.findAll("div",
                         {"class": "caja_contenedor_masvistos_modulo_texto"})

# Obtenemos las URLs de aquellas noticias
urlViewed = []


for div in mostViewed:
    if div.a:
        urlViewed.append(div.a["href"])


# Para cada una de esas noticias ingresamos a su página
# y obtenemos su título y su contenido.
count = 1

# Las siguientes dos líneas crean el archivo (mode="w") y escriben la primera
# columna, que es el encabezado del título. En python, para escribir un TAB
# se utiliza \t y para escribir un salto de línea (ENTER) se usa \n
with open("noticias.tsv", mode="w") as file:
  file.write("número\ttítulo\tcontenido\n")

for url in urlViewed:
    try:
        page = urlopen(
            "http://www.emol.com{}".format(url), context=context
        ).read()
        soup = BeautifulSoup(page)
    except:
        print("Error al intentar recuperar noticia.")

    try:
        title_h1 = soup.find("h1", {"id": "cuDetalle_cuTitular_tituloNoticia"})
        title = title_h1.getText()
    except:
        title = "Error al intentar recuperar título de noticia."

    try:
        content_div = soup.find("div", {"id": "cuDetalle_cuTexto_textoNoticia"})
        content = content_div.getText()
    except:
        content = "Error al intentar recuperar contenido de noticia."
    
    
    # Aquí volvemos a abrir el archivo, pero en modo "a", esto es de append
    # Es decir, abre el archivo y nos situamos en la úlima línea para agregar 
    # más información. Si abren el archivo en modo "w" (write), van a perder lo que tenían
    # antes.
    
    with open("noticias.tsv", mode="a") as file:
      file.write("{}\t{}\t{}\n".format(count, title, content))
      
    print("Noticia {} lista!".format(count))
    
    count += 1
```

### Leer tweets descargados


```python
import json
def json_readr(file):
    for line in open(file, mode="r"):
        yield json.loads(line)
```

```python
dataj = json_readr("tweets.jsonl")
```

```python
dataj_lst = list(dataj)
print(len(dataj_lst))
```

```python
tweets = list()
for tweet in dataj_lst:
    if 'retweeted_status' not in tweet:
        tweets.append((tweet['created_at'], tweet['id_str'], tweet['full_text'], tweet['user']['id_str'], tweet['user']['name'], tweet['user']['screen_name']))
        #print(tweet)
        #break
```

```python
len(tweets)
```

```python
df = pd.DataFrame(tweets, columns = ['created_at', 'id_tweet', 'text', 'id_user', 'name', 'screen_name'])
```

```python
# guardar los datos

df.to_csv("tweets_search.csv",
                          sep = ";", 
                         encoding = "UTF-8",
                         quotechar = "\"",
                         index = False,
                        quoting=csv.QUOTE_NONNUMERIC)
```
